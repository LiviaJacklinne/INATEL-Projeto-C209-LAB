```python
# Importando a biblioteca cv2, para usarmos OpenCV
import cv2
from matplotlib import pyplot as plt 

# abrindo a imagem
imgx = cv2.imread("image.jpg") 
  
# Passando as imagens para cinza e RGB
imgx_gray = cv2.cvtColor(imgx, cv2.COLOR_BGR2GRAY) 
imgx_rgb = cv2.cvtColor(imgx, cv2.COLOR_BGR2RGB) 
 
stop_data = cv2.CascadeClassifier('stop_data.xml') 
  
found = stop_data.detectMultiScale(imgx_gray, minSize =(20, 20)) 

amount_found = len(found) 
  
if amount_found != 0: 
    for (x, y, width, height) in found: 
        
        cv2.rectangle(imgx_rgb, (x, y),  
                (x + height, y + width),  
                
                      (0, 255, 0), 10) 
            
          
    plt.subplot(1, 1, 1) 
plt.imshow(imgx_rgb) 
plt.show() 
```

```python
# abrindo a imagem
imgy = cv2.imread("aux.jpg") 
  
# Passando as imagens para cinza e RGB
imgy_gray = cv2.cvtColor(imgy, cv2.COLOR_BGR2GRAY) 
imgy_rgb = cv2.cvtColor(imgy, cv2.COLOR_BGR2RGB) 
  
stop1_data = cv2.CascadeClassifier('stop_data.xml') 
  
found = stop1_data.detectMultiScale(imgy_gray, minSize =(20, 20)) 

amount_found = len(found) 
  
if amount_found != 0: 
    for (x, y, width, height) in found: 
        
        cv2.rectangle(imgy_rgb, (x, y),  
                (x + height, y + width),  
                
                      (255, 255, 255), 25) 
            
          
    plt.subplot(1, 1, 1) 
plt.imshow(imgy_rgb) 
plt.show() 
```
